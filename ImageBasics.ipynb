```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Drawing Shapes on the blank image
*   Line
*   Rectangle
*   Square
*   Circle
*   Polygon
*   Text







```python
img_blank = np.zeros((512, 512, 3), dtype=np.int16)
img_blank
```

```python
plt.imshow(img_blank)
```

```python
cv2.rectangle(img=img_blank, pt1=(100, 200), pt2=(300,300), color=(255, 0, 0), thickness=2)
```

```python
cv2.rectangle(img=img_blank, pt1=(100, 200), pt2=(300,300), color=(255, 100, 200), thickness=2)
plt.imshow(img_blank)
```

```python
cv2.rectangle(img=img_blank, pt1=(100, 100), pt2=(200,200), color=(255, 200, 100), thickness=2)
plt.imshow(img_blank)
```

```python
cv2.circle(img=img_blank, center=(250, 150), radius=50, color=(155, 200, 100), thickness=2)
plt.imshow(img_blank)
```

```python
cv2.circle(img=img_blank, center=(350, 250), radius=50, color=(155, 250, 100), thickness=-1)
plt.imshow(img_blank)
```

```python
cv2.line(img=img_blank, pt1=(100, 300), pt2=(250, 150), color=(255, 200, 100), thickness=2)
cv2.line(img=img_blank, pt1=(100, 200), pt2=(200,100), color=(105, 100, 200), thickness=3)
plt.imshow(img_blank)
```

```python
cv2.putText(img=img_blank, text='Hlo Aks', org=(10, 90), fontFace=cv2.FONT_HERSHEY_DUPLEX, fontScale=3, color=(110, 110, 110), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(img_blank)
```

```python
vertices = np.array([ [100, 400], [250, 325], [400, 400], [300, 450] ], dtype=np.int32)
vertices
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1, 1, 2))
pts
```

```python
pts.shape
```

```python
cv2.polylines(img=img_blank, pts=[pts], isClosed=True, color=(255, 0, 0), thickness=2)
plt.imshow(img_blank)
```

# Draw the circle using the Mouse click

### This shell will work in the jupyter notebook

*   No GUI backend by Google Colab
*   There's no window system to capture mouse events



```python
def draw_circle(event, x, y, flags, param):
  if event == cv2.EVENT_LBUTTONDOWN:
    cv2.circle(img=img, center=(x, y), radius=50, color=(255, 0, 0), thickness=-1)

cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing', draw_circle)

img = np.zeros((512, 512, 3), dtype=np.int8)
while True:
  cv2.imshow('my_drawing', img)
  if cv2.waitKey(20) & 0xFF == 27:
    break

cv2.destroyAllWindows()
```
